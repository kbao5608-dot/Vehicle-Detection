```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolov8_detrac.pt")
```

```python
# img_path = "./dataset/valid/images/009981_jpeg.rf.c7f939308fe682f59478636b1d4fe341.jpg"
img_path = "./test_img/test4.jpg"
ans = model.predict(img_path, imgsz=640, conf=0.1)

bbox = ans[0].plot()
plt.imshow(cv2.cvtColor(bbox, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
